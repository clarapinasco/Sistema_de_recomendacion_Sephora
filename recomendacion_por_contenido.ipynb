```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
product_info =pd.read_csv("data/product_info.csv")
product_info = product_info[["product_id", "product_name"]]
```

```python
df = pd.read_csv("data/df_pca.csv")
```

Hacemos un df para poder conseguir el nombre del producto a través de su identificación única.

```python
nombres = df[["product_id"]].merge(product_info, on='product_id', how='left')
```

```python
df.head()
```

```python
# Función para encontrar productos similares basados en el coseno de similitud
def find_similar_products(product_id, num_similar_products=3):
    product_index = df.loc[df['product_id'] == product_id].index[0]
    # Extraer los componentes principales
    principal_components = df.drop(columns=['product_id']).values
    # Calcular la matriz de similitud de coseno
    cosine_sim = cosine_similarity(principal_components, principal_components)
    # Obtener las similitudes del producto dado
    product_similarities = cosine_sim[product_index]
    # Ordenar y obtener los índices de los productos más similares
    similar_products_indices = product_similarities.argsort()[::-1][1:num_similar_products+1]
    # Obtener los productos similares
    similar_products = df.loc[similar_products_indices, ['product_id']]
    return similar_products

# Verificar si el ID del producto ingresado existe en el dataset
try:
    similar_products = find_similar_products('P107306', num_similar_products=3)
    print("Producto buscado:")
    print(nombres[nombres["product_id"]== "P107306"])
    print('Productos similares:')
    for i, product in similar_products.iterrows():
        print(f'Product ID: {product.product_id}')
        print(nombres[nombres["product_id"]== product.product_id]["product_name"])
        print("--------------------------------------")
except ValueError:
    print('Disculpas, no pudimos encontrar ese ID de producto. Por favor ingresa otro.')
except IndexError:
    print('Disculpas, no pudimos encontrar ese ID de producto. Por favor ingresa otro.')
```
